```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('../data/rankings/archipelago_ROUND4.json', 'rb') as f:
    data = json.load(f)
```

```python
my_profit = [i['profit'] for i in data if i['team']['name'] == 'Arbland'][0]
```

```python
profits = np.array([i['profit'] for i in data])
```

```python
plt.figure(figsize=(6, 3))
plt.hist(profits[:4100], bins=50)
plt.axvline(my_profit, c='r')
plt.show()
```

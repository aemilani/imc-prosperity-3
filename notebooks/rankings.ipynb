```python
import json
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('../data/rankings/archipelago_ROUND6.json', 'rb') as f:
    data = json.load(f)
```

```python
my_profit = [i['profit'] for i in data if i['team']['name'] == 'Arbland'][0]
```

```python
profits = np.array([i['profit'] for i in data])
profits = np.sort(profits)[::-1]
```

```python
fig, ax = plt.subplots(figsize=(6, 3))
sns.kdeplot(profits[:13500], bw_adjust=1, ax=ax)  # you can adjust smoothness with bw_adjust
plt.axvline(my_profit, c='r')
plt.xlabel('Profit')
plt.ylabel('Density')
plt.show()
```

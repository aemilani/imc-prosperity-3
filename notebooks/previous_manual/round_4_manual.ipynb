```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
```

```python
def b1_opt(b2_avg):
    if b2_avg <= 978:
        return 952
    else:
        return (900 + 2 * b2_avg) / 3
```

```python
def opt_b1_expected_return(b2, b2_avg):
    b1 = b1_opt(b2_avg)
    if b2_avg <= b2:
        e = (1/10000*b1**2 - 9/50*b1 + 81)*(1000 - b1) + (1/10000*b2**2 - 9/50*b2 - 1/10000*b1**2 + 9/50*b1)*(1000-b2)
        return e
    else:
        e = (1/10000*b1**2 - 9/50*b1 + 81)*(1000 - b1) + (1/10000*b2**2 - 9/50*b2 - 1/10000*b1**2 + 9/50*b1)*(1000-b2_avg)
        return e
```

```python
b2_avg = 970

plt.figure(figsize=(4, 2))
plt.plot(range(int(b1_opt(b2_avg)), 1000), [opt_b1_expected_return(b2, b2_avg) for b2 in range(int(b1_opt(b2_avg)), 1000)])
plt.axvline(b2_avg, c='r')
plt.show()
```

```python
def opt_expected_return(b2_avg):
    if b2_avg <= 978:
        return 20.4152
    else:
        b1 = b1_opt(b2_avg)
        b2 = b2_avg
        e = (1/10000*b1**2 - 9/50*b1 + 81)*(1000 - b1) + (1/10000*b2**2 - 9/50*b2 - 1/10000*b1**2 + 9/50*b1)*(1000-b2_avg)
        return e
```

```python
plt.figure(figsize=(4, 2))
plt.plot(range(900, 1000), [opt_expected_return(b2_avg) for b2_avg in range(900, 1000)])
plt.show()
```

```python
def opt_bids(b2_avg):
    if b2_avg <= 978:
        return 952, 978
    else:
        return b1_opt(b2_avg), b2_avg
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 3))
ax[0].plot(range(970, 1000), [opt_bids(b2_avg)[0] for b2_avg in range(970, 1000)])
ax[0].set_xlabel('Average b2')
ax[0].set_ylabel('Optimal b1')
ax[1].plot(range(970, 1000), [opt_bids(b2_avg)[1] for b2_avg in range(970, 1000)])
ax[1].set_xlabel('Average b2')
ax[1].set_ylabel('Optimal b2')
plt.show()
```

```python
for b2_avg in range(978, 990):
    b1, b2 = opt_bids(b2_avg)
    print(f'{b2_avg}: {int(np.round(b1)), b2}')
```

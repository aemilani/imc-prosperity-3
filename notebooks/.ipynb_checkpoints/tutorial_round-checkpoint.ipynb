```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = pd.read_csv('../data/round_0/prices_round_0_day_-2.csv', delimiter=';').drop(columns=['day', 'profit_and_loss'])
prices_resin = prices[prices['product'] == 'RAINFOREST_RESIN']
prices_kelp = prices[prices['product'] == 'KELP']
```

```python
plt.figure(figsize=(6, 3))
plt.plot(prices_resin['mid_price'])
plt.title('Resin Price')
plt.xlabel('timestamp')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(prices_kelp['mid_price'])
plt.title('Kelp Price')
plt.xlabel('timestamp')
plt.ylabel('price')
plt.show()
```
